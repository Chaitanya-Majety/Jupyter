```python
import math
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

x = [2,3,1,1,2,4,8,10,11,20,21,15,100,110,200]

mu = sum(x)/len(x)

sigma_nume = 0

for i in x:
    sigma_nume += ((i-mu)*(i-mu))

sigma = math.sqrt(sigma_nume/len(x))

l = sorted(x)

if len(l)%2 ==0:
    median = l[int(len(l)/2)-1]
else:
    median = l[math.ceil(len(x)/2)-1]+l[math.floor(len(x)/2)-1]

xz = []

for i in x:
    xz.append(((i-mu)/sigma))

mu_new = sum(xz)/len(xz)

sigma_nume_new = 0

for i in xz:
    sigma_nume_new += ((i-mu_new)*(i-mu_new))

sigma_new = math.sqrt(sigma_nume_new/len(xz))

l = sorted(xz)

if len(l)%2 ==0:
    median_new = l[int(len(l)/2)-1]
else:
    median_new = l[int(math.ceil(len(x)/2))-1]+l[int(math.floor(len(x)/2))-1]

print(mu,sigma,median,"\n",mu_new,sigma_new,median_new)

data = {"a":x,"b":xz}

df = pd.DataFrame(data["a"])
df1 = pd.DataFrame(data["b"])
print(x,xz)
print(df.plot.kde(),"\t",df1.plot.kde())
```
