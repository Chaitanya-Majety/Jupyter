```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
%matplotlib inline
```

```python
data = pd.read_csv("FitBit data.csv")
data.head()
```

```python
data1 = data.copy()
data1["ActivityDate"] = pd.to_datetime(data1["ActivityDate"])
data1["year"] = data1["ActivityDate"].map(lambda x: x.year)
data1["month"] = data1["ActivityDate"].map(lambda x: x.month)
data1["date"] = data1["ActivityDate"].map(lambda x: x.day)
data1.head()
```

```python
data1.drop("TrackerDistance",axis=1,inplace=True)
data1.head(200)
```

```python
plt.figure(figsize=(25,10))

ax = sns.boxplot(x="date",y="Calories",data=data1)
ax1 = sns.swarmplot(x="date",y="Calories",data=data1)
```

```python
data1["week"] = data1["ActivityDate"].dt.week
data1["ActivityDate"] = pd.to_datetime(data1["ActivityDate"])
data1["day"]= data1["ActivityDate"].map(lambda x: pd.Timestamp(x).day_name())
```

```python
plt.figure(figsize=(25,10))
ax = sns.barplot(x="day",y="Calories",data=data1) ## those black lines are outliers
```

```python
ax = sns.scatterplot(x="Calories",y="SedentaryMinutes",data=data1)
```
